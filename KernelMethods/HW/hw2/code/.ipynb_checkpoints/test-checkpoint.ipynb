```python
import numpy as np
import pickle as pkl

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
# Import the KernelPCA class implementation

class KernelPCA:    
    def __init__(self, kernel, r=2, center=False):                             
        self.kernel = kernel
        self.alpha = None  # Matrix of shape N times d representing the d eigenvectors alpha
        self.lmbda = None  # Vector of size d representing the top d eigenvalues
        self.support = None  # Data points where the features are evaluated
        self.r = r  # Number of principal components
        self.center = center
    
    def compute_PCA(self, X):
        self.support = X
        K = self.kernel(X, X)
        if self.center:
            N = K.shape[0]
            ones = np.ones((N,N)) / N
            K = K - ones.dot(K) - K.dot(ones) + ones.dot(K).dot(ones)  # Center the kernel matrix
        eigvalues, eigvectors = np.linalg.eigh(K)
        idx = np.argsort(eigvalues)[::-1]
        self.lmbda = eigvalues[idx][:self.r]
        self.alpha = eigvectors[:,idx][:,:self.r] / np.sqrt(self.lmbda)
        plt.plot(self.lmbda)  # Plotting eigenvalues (optional)
        
    def transform(self, x):
        G = self.kernel(x, self.support)
        if self.center:
            N = G.shape[0]
            one_n = np.ones((N,N)) / N
            G = G - one_n.dot(G) - G.dot(one_n) + one_n.dot(G).dot(one_n)  # Center the kernel matrix
        return np.dot(G, self.alpha)

# Example usage with make_circles dataset
X, y = make_circles(n_samples=1000, noise=0.1, factor=0.2, random_state=42)

# Define a Gaussian kernel function
def gaussian_kernel(X, Y, sigma=1.0):
    pairwise_dists = np.sum(X**2, axis=1)[:, np.newaxis] + np.sum(Y**2, axis=1) - 2 * np.dot(X, Y.T)
    return np.exp(-pairwise_dists / (2 * sigma ** 2))
```

```python
file = open('datasets/pca_datasets', 'rb')
datasets = pkl.load(file)
file.close()
X,  y = datasets['dataset_1']['train']['x'], datasets['dataset_1']['train']['y']
test_set = datasets['dataset_1']['test']
```

```python
# Initialize and compute Kernel PCA
kpca = KernelPCA(kernel=gaussian_kernel, r=2, center=True)
kpca.compute_PCA(X)
transformed_data = kpca.transform(X)

# Plot original and transformed data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Data')
plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.title('Transformed Data')
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=y, marker='o', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```
